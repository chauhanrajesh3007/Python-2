# Data Cleaning

# -->remove null values

```python
import pandas as pd
df=pd.read_csv('sales_data.csv')
print(df)
```

```python
print(df.isna())#df.isnull()-->give true or false
```

```python
print(df.isna().sum())#-->give colum wise count
```

```python
print(df.dropna())#-->remove null value rows
```

```python
print(df.dropna(axis=1))#inplace=True -->remove null value as columswise
```

```python
print(df.dropna(thresh=2))#-->row wise null value is greater than 2 then remove this row
```

```python
print(df.dropna(how='all'))#--> remove this row whose all value is null
```

```python
print(df.dropna(how='any'))#--> remove this row whose anyone value is null
```

```python
print(df.dropna(subset=['sales','expenses']))#-->remove this row sales and expenses value is null
```

# -->fill null value

```python
print(df['sales'].fillna(0))#-->fill null value in sales colum as 0
```

```python
print(df['sales'].fillna(df['sales'].mean()))#-->fill null value in sales colum as mean of sales
```

```python
print(df.fillna(1))#-->fill all null value as 1
```

```python
df=pd.read_csv("auto-mpg.csv")
q1=df["acceleration"].quantile(0.25)
q3=df["acceleration"].quantile(0.75)
IQR=q3-q1
high=q3+(IQR*1.5)
low=q1-(IQR*1.5)
print(high,low)
df.loc[(df["acceleration"]<low) | (df["acceleration"]>high)]
```

```python
df=df.loc[(df["acceleration"]>=low) & (df["acceleration"]<=high)]
df
```
